```python
import pandas as pd

df1 = pd.read_csv("hf://datasets/azrai99/coursera-course-dataset/coursera_course_2024.csv")
df1.head()
```

```python
df1.drop(columns=['Unnamed: 0','Satisfaction Rate'], inplace=True)
df1.head()
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
mean_num_reviews = df1['num_reviews'].mean()
df1['num_reviews'].fillna(mean_num_reviews, inplace=True)
```

```python
import re

df1['Modules/Courses'] = df1['Modules/Courses'].str.extract('(\d+)').astype(float)
mode_modules_courses = df1['Modules/Courses'].mode()[0]
df1['Modules/Courses'].fillna(mode_modules_courses, inplace=True)
df1['Modules/Courses'] = df1['Modules/Courses'].astype(int).astype(str)
```

```python
df1.head()
```

```python
df1.drop(columns=['modules'], inplace=True)
df1.head()
df1.isnull().sum()
```

```python
df1['Description'] = df1.apply(lambda row: "In this course we are going to deep dive in to the topic " + row['title'] if pd.isnull(row['Description']) else row['Description'], axis=1)
```

```python
mode_level = df1['Level'].mode()[0]
df1['Level'].fillna(mode_level, inplace=True)
```

```python
mode_level = df1['Instructor'].mode()[0]
df1['Instructor'].fillna(mode_level, inplace=True)
```

```python
import re

df1['Schedule'] = df1['Schedule'].str.extract('(\d+)').astype(float)
mean_schedule = df1['Schedule'].mean()
df1['Schedule'].fillna(mean_schedule, inplace=True)
df1['Schedule'] = df1['Schedule'].astype(int).astype(str) + ' hours to complete'
```

```python
df1.isnull().sum()
```

```python
numerical_cols = ['rating', 'enrolled', 'num_reviews', 'Modules/Courses', 'Schedule']

for col in numerical_cols:
  df1[col] = pd.to_numeric(df1[col], errors='coerce')

categorical_cols = ['Level', 'Organization']

for col in categorical_cols:
    df1[col] = pd.Categorical(df1[col])
```

```python
import numpy as np

numerical_cols = ['rating', 'num_reviews', 'Modules/Courses']

for col in numerical_cols:
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df1[(df1[col] < lower_bound) | (df1[col] > upper_bound)]
    print(f"Outliers in {col}:\n{outliers}\n")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in numerical_cols:
    plt.figure(figsize=(8, 6))

    data_to_plot = pd.to_numeric(df1[col], errors='coerce').dropna()
    sns.boxplot(x=data_to_plot)  # Use the cleaned data for plotting
    plt.title(f'Box Plot of {col} showing Outliers')
    plt.xlabel(col)
    plt.show()
```

```python

sns.histplot(df1['enrolled'])
plt.title('Distribution of enrolled')
plt.show()
```

```python


top_20 = df1.sort_values(by=['rating'], ascending=False).head(20)  # Select top 20 by rating


highest_rated_courses = top_20[['Instructor', 'title', 'rating']]

plt.figure(figsize=(10, 8))  # Adjust figure size for better visibility
sns.barplot(x='rating', y='title', hue='Instructor', data=highest_rated_courses, orient='h', dodge=False)
plt.title('Top 20 Highest Rated Courses')
plt.xlabel('Rating')
plt.ylabel('Course')
plt.legend(title='Instructor', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print(highest_rated_courses) # Print details of top 20 courses
```
