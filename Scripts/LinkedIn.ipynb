```python
embedding_cols = [col for col in df.columns if col.startswith('embeddings_')]
df = df.drop(columns=embedding_cols)
```

```python
df.head()
```

```python
df = df.drop(columns=['About Me' ,'Recommendations', 'text', 'Location', 'Education'])
df.head(20)
```

```python
df.rename(columns={'Headline': 'Current Role'}, inplace=True)
df.head()
```

```python
df['Current Role'] = df['Current Role'].str.split('|', n=1, expand=True)[0]
```

```python
df['Contact mail'] = df['FirstName'].str.lower() + '.' + df['LastName'].str.lower() + '@outlook.com'
```

```python
df.head()
```

```python
data_scientists = df[df['Current Role'].str.contains('data scientist', case=False, na=False)]
data_scientists = df[df['Current Role'].str.contains('data scientist', case=False, na=False)]
data_scientists.head(10)
```

```python
all_skills = []
for skills in data_scientists['Skills']:
    if isinstance(skills, str):  # Check if the value is a string
        all_skills.extend(skills.split(';'))  # Split by ';'
```

```python
from collections import Counter

all_skills = []
for skills in data_scientists['Skills']:
    if isinstance(skills, str):
        all_skills.extend(skills.split(';'))

skill_counts = Counter(all_skills)
```

```python
import matplotlib.pyplot as plt

# Get the most common skills and their frequencies
top_skills = skill_counts.most_common(20)
skill_names, frequencies = zip(*top_skills)

plt.figure(figsize=(10, 6))
plt.bar(skill_names, frequencies)
plt.xlabel('Skills')
plt.ylabel('Frequency')
plt.title('Top Skills for Data Scientists')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

all_roles = df['Current Role'].str.strip().tolist()

# Convert to lowercase and split into words
all_roles = [' '.join(role.lower().split()) for role in all_roles if isinstance(role, str)]

```

```python
from collections import Counter

role_counts = Counter(all_roles)
```

```python
top_ten_roles = role_counts.most_common(20)
for role, count in top_ten_roles:
    print(f"{role}: {count}")
```

```python
import matplotlib.pyplot as plt

roles, counts = zip(*top_ten_roles)

plt.figure(figsize=(12, 6))
plt.bar(roles, counts)
plt.xlabel('Current Roles')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common Current Roles')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```
